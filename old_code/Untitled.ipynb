```python
import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create contour plot
plt.contour(X, Y, Z)
plt.colorbar()  # Add a colorbar to the plot
plt.title('Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some data
x = np.linspace(-1000, 1000, 1000)
y = np.linspace(-1000, 1000, 1000)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create the contour plot
plt.contourf(X, Y, Z)
plt.colorbar()  # Add a colorbar
plt.title('Continuous Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
